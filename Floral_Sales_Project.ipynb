# **Combining Dataframes Together**

```python
#Concatenating the dataframes from each year into one dataframe
fl_df = pd.concat([df18,df19,df20,df21,df22,df23])
print(fl_df)
```

```python
#Looking at a preview of the new dataframe to ensure it was concatenated correctly
fl_df.head()
```

```python
fl_df = fl_df[fl_df['Order #'].str.isnumeric()]
fl_df['Order #'] = fl_df['Order #'].astype(int)
```

# Website Orders CSV

```python
#Importing the web order data
wo = pd.read_csv("C:/Users/davyd/OneDrive/Desktop/website_orders_january_1_2019-june_6_2023.csv")
wo.head()
```

```python
wo.rename(columns={'Order Number':'Order #'},inplace = True)
```

```python
#Merging the two dataframes together so everything is in one place
joined_data = pd.merge(fl_df, wo, on='Order #', how='outer')
print("Joined Data:")
print(joined_data)
```
